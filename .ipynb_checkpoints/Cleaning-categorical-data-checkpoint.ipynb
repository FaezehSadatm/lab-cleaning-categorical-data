#### 5. Plot every categorical variable. What can you see in the plots? Note that in the previous lab you used a bar plot to plot categorical data, with each unique category in the column on the x-axis and an appropriate measure on the y-axis. However, this time you will try a different plot. This time in each plot for the categorical variable you will have, each unique category in the column on the x-axis and the target(which is numerical) on the Y-axis.

```python
for column in categorical_df.columns:
    plt.figure(figsize=(12,6))
    sns.barplot(x=column, y="total_claim_amount", data=customer_df)
    plt.show()
```

#### 6. For the categorical data, check if there is any data cleaning that need to perform. Hint: You can use the function value_counts() on each of the categorical columns and check the representation of different categories in each column. Discuss if this information might in some way be used for data cleaning.

```python
for column in categorical_df.columns:
    print(customer_df[column].value_counts())
```

```python
customer_df['employmentstatus'].value_counts()
```

```python
customer_df['employmentstatus'] = customer_df['employmentstatus'].apply(lambda x: x if x in ['Employed', 'Unemployed'] else 'Other')

customer_df['employmentstatus'].value_counts()
```

```python
customer_df['policy'].value_counts()
```

```python
customer_df['policy'] = customer_df['policy'].apply(lambda x: 'Special' if x in ['Special L3','Special L2','Special L1'] else x)

customer_df['policy'].value_counts()
```

```python
customer_df['vehicle_class'].value_counts()
```

```python
customer_df['vehicle_class'] = customer_df['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Luxury SUV', 'Luxury Car'] else x)

customer_df['vehicle_class'].value_counts()
```

#### My comment: as we can see we have too many categories in 'employmentstatus', 'policy' and  'vehicle_class' columns which some are little represented so we can bucket several of them into one.
